```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# reading the file of heart patients

```python
heart_df=pd.read_csv('C:\\Users\\HP\\Downloads\\heart.csv')
pd.set_option('display.max_rows',None)
heart_df
```

# cleaning the data set

```python
heart_df=heart_df.interpolate(how='index')
heart_df
```

# Peoples age more than 40 years having heart disease

```python
heart_df[heart_df['Age']>40].sort_values(by=['Age'])
```

# Number of males having heart disease

```python
 heart_df[heart_df['Sex']=='M']
```

# Number of females having heart disease

```python
heart_df[heart_df['Sex']=='F']
```

# Graph of heart diseases age vs sex

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Sex',y='Age',data=heart_df)
plt.show()
```

# Cholestrol Graph

```python
plt.figure(figsize=(7,5))
sns.barplot(x='Sex',y='Cholesterol',data=heart_df,edgecolor='m',linewidth=4)
plt.title('Cholestrol for male and female',fontsize=19)
plt.xlabel('Male and Female',color='c',fontsize=14)
plt.ylabel('Cholestrol data',fontsize=14,color='c')
plt.show()
```

# Scatter graph between cholestrol and chest pain

```python
plt.figure(figsize=(9,7))
sns.scatterplot(x='ChestPainType',y='Cholesterol',data=heart_df,hue='Sex',size='Sex',sizes=(100,200),marker='o')
plt.title('Cholestrol-chestpain graph between male and female',fontsize=20,color='m')
plt.show()
```

# Peoples having resting EGC normal

```python
heart_df[heart_df['RestingECG']=='Normal']
```

# Total people suffering from heart disease 

```python
tot=heart_df[heart_df['HeartDisease']==0].value_counts().sum()
totf=heart_df[heart_df['HeartDisease']==1].value_counts().sum()
print('Number of peoples not suffering from heartdiesease are:',tot)
print('Number of peoples suffering from heartdiesease are:',totf)
```

# Graph for Males and Females

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='Sex',y='MaxHR',data=heart_df,color='g',palette='hot',label='Rising-rate')
plt.title('Heart-rate',fontsize=14,color='m')
plt.xlabel('Sex',fontsize=14,color='r')
sns.set()
plt.ylabel('Max-heartrate',fontsize=14,color='r')
plt.show()
```
